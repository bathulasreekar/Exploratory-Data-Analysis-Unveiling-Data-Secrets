```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

# Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MM = MinMaxScaler()
```

```python
df1['MMTB'] = MM.fit_transform(df1[['total_bill']])
df1.head()
```

```python
print("Skewness of Original Total Bill",df1['total_bill'].skew())
print("Skewness of Min Maxed Total Bill",df1['MMTB'].skew())
```

```python
print("Kurtosis of Original Total Bill",df1['total_bill'].kurt())
print("Kurtosis of Min Maxed Total Bill",df1['MMTB'].kurt())
```

```python
sns.kdeplot(df1['total_bill'])
plt.show()
sns.kdeplot(df1['MMTB'])
plt.show()
```

# Transformation using SQRT & LOG

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
df1['SQRT TB'] = np.sqrt(df1['total_bill'])
df1['LOG TB'] = np.log(df1['total_bill'])
```

```python
df1.head()
```

```python
print('Skewness of Original Total Bill',df1['total_bill'].skew())
print('Skewness of SQRT Total Bill',df1['SQRT TB'].skew())
print('Skewness of LOG Total Bill',df1['LOG TB'].skew())
```

```python
print('Kurtosis of Original Total Bill',df1['total_bill'].kurt())
print('Kurtosis of SQRT Total Bill',df1['SQRT TB'].kurt())
print('Kurtosis of LOG Total Bill',df1['LOG TB'].kurt())
```

```python
sns.kdeplot(df1['total_bill'])
plt.show()
sns.kdeplot(df1['SQRT TB'])
plt.show()
sns.kdeplot(df1['LOG TB'])
plt.show()
```
